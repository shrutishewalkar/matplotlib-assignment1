Que: what is a scatter plot? use the following code to generate data for x, y . using this generated data plot as scatter plot. 

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my scatter data")
```

Que: why is subplot() function used? draw four line plot using subplot fun

subplots() which acts as a utility wapper and helps in creating common layouts of subplots, including the encloseing figure object , in a single call

```python
import matplotlib.pyplot as plt
import numpy as np
x=([0,1,2,3,4,5])
y1=([0,100,200,300,400,500])
y2=([50,20,40,20,60,70])
y3=([10,20,30,40,50,60])
y4=([200,350,250,550,450,150])

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)


plt.show()
```

Que : what is the box plot? why it is used? using the following data plot a box plot

box plots are used to show dis tributions of numberis data values, espically when you want to compare them between multiple groups.. A box plot is a away to visualize the distribution of the data by using box and some vertical lines

```python
import matplotlib.pyplot as plt
import numpy as np

spread_1 = np.random.normal(100,10,200)
spread_2 = np.random.normal(90,20,200)
all_spread = [ spread_1 , spread_2]
plt.boxplot(all_spread, labels=['spread_1', 'spread_2'])
plt.show()
```
